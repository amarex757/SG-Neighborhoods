```python
import pandas as pd

import json # library to handle JSON files
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
import pandas.io.json._normalize # tranform JSON file into a pandas dataframe
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans
import folium # map rendering library

print('Libraries imported.')
```

```python
with open('buildings.json') as json_data:
    singapore_data = json.load(json_data)
```

```python
singapore_data
```

```python
neighborhoods_data = singapore_data[0]
```

```python
neighborhoods_data['BUILDING']
```

```python
columns = ['Postcode', 'Neighborhood', 'Latitude', 'Longitude']
neighborhoods = pd.DataFrame(columns=columns)
```

```python
neighborhoods
```

```python
for data in singapore_data:
    post= data['POSTAL']
    neighborhood_name = data['BUILDING']    
    neighborhood_lat = data['LATITUDE']   
    neighborhood_long = data['LONGITUDE']    
    neighborhoods = neighborhoods._append({'postcode': post, 'neighborhood': neighborhood_name, 'latitude': neighborhood_lat, 'longitude': neighborhood_long}, ignore_index=True) 
```

```python
neighborhoods.head()
```

```python
neighborhoods.to_csv('singapore_geo.csv', encoding='utf-8')
```

## 1. Cleaning data
#### Remove all NIL values in Neighborhood


```python
neighborhoods = neighborhoods[neighborhoods['Neighborhood']!='NIL']
```

```python
neighborhoods.shape
```

```python
neighborhoods.info()
```

```python
neighborhoods['Latitude'] = neighborhoods['Latitude'].map(lambda x: float(x))
neighborhoods['Longitude'] = neighborhoods['Longitude'].map(lambda x: float(x))
```

```python
neighborhoods.to_csv('singapore_geo.csv', encoding='utf-8')
```
